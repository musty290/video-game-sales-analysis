```python
#  Video Game Sales EDA & Market Insights

# ## Objective
# Analyze global video game sales to identify:
# - Top-performing genres and publishers
# - Sales trends over time
# - Regional differences
# - Actionable business insights
# ###
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:/Users/USER/Downloads/vgsales.csv")
data.head()
```

```python
data.shape
data.info()
data.isna().sum()
```

```python
# Drop missing values
data_clean = data.dropna(subset=['Year','Publisher'])

# Convert types
data_clean['Year'] = data_clean['Year'].astype(int)

# Remove duplicates
data_clean = data_clean.drop_duplicates()
data_clean
```

```python
# Top genres by sales
genre_sales = data_clean.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

# Publisher performance
publisher_sales = data_clean.groupby('Publisher')['Global_Sales'].mean()

# Yearly trends
yearly_sales = data_clean.groupby('Year')['Global_Sales'].sum().reset_index()
yearly_sales['Growth'] = yearly_sales['Global_Sales'].pct_change()
```

```python
sns.countplot(data=data_clean, x='Genre')
plt.xticks(rotation=45)
plt.show()

sns.histplot(data_clean['Global_Sales'], bins=15)
plt.show()

sns.boxplot(x=data_clean['Global_Sales'])
plt.show()
```

```python
sns.scatterplot(data=data_clean, x='Rank', y='Global_Sales')
plt.show()

sns.lineplot(data=data_clean, x='Year', y='Global_Sales')
plt.show()

sns.boxplot(data=data_clean, x='Genre', y='Global_Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.heatmap(data_clean.corr(numeric_only=True), annot=True)
plt.show()

sns.scatterplot(data=data_clean, x='NA_Sales', y='EU_Sales', hue='Genre')
plt.show()
```

```python
# ## Key Insights

# - Action and Sports genres dominate global sales.
# - Sales are highly skewed — few games generate most revenue.
# - Strong correlation between NA and EU sales.
# - Some genres show declining trends over time.
```
